### The objective of this project is to conduct a comprehensive time series analysis of the stock prices of selected companies from the National Stock Exchange (Reliance, TCS, Infosys, Britannia, Nestle India, Vedanta Limited, Sunpharma, Dr Reddy, Bharti Airtel and HDFC Bank). 

### The analysis aims to uncover insights, trends, and potential trading signals based on historical data, and utilize the Prophet forecasting model to predict future stock prices. 

```python
import pandas as pd
from prophet import Prophet
import yfinance as yf
from datetime import datetime, time, timedelta
import pytz
import matplotlib.pyplot as plt
```

```python
#Define the list of stock tickers to analyse
stock_tickers=['RELIANCE.NS','TCS.NS','INFY.NS','BRITANNIA.NS','NESTLEIND.NS','VEDL.NS','SUNPHARMA.NS','DRREDDY.NS','BHARTIARTL.NS','HDFCBANK.NS']

#Define the start and end dates of the historical data
start_date='2018-05-11' #5 years from now
end_date='2023-05-11' #current date
```

```python
#Get IST Timezone
ist_timezone=pytz.timezone('Asia/Kolkata')
```

```python
#Get current date and time in IST
curr_datetime=datetime.now(ist_timezone)
print(curr_datetime)
```

```python
#Check if current time is before 3:30pm IST, and if so, subtract one day from current date
if curr_datetime.time() < time(hour=15, minute=30):
    curr_datetime -= timedelta(days=1)
```

```python
#Extract current date from current datetime
curr_date=curr_datetime.date()
print(curr_date)
```

```python
#Extract current time from current datetime
curr_time=curr_datetime.time()
print(curr_time)
```

```python
#Loop through each stock ticker and analyse the data
for ticker in stock_tickers:
    #Download distorical data from yFinance
    stock_data=yf.download(ticker, start=start_date, end=end_date, interval='1d')
    #Calculate 100 and 200 days Simple Moving Averages
    stock_data['SMA100']=stock_data['Close'].rolling(window=100).mean()
    stock_data['SMA200']=stock_data['Close'].rolling(window=200).mean()
    
    # Filter data up to current date
    stock_data = stock_data[stock_data.index <= curr_datetime.strftime('%Y-%m-%d')]
      
    latest_data=stock_data.iloc[-1]
    #Check if closing price is greater than both 100 and 200 day moving averages
    if latest_data['Close']>latest_data['SMA100'] and latest_data['Close']>latest_data['SMA200']:
        print(f"Buy {ticker} stock: Closing price is above 100 and 200 day SMA")
    #Check if closing price is less than both 100 and 200 day moving averages
    elif latest_data['Close']<latest_data['SMA100'] and latest_data['Close']<latest_data['SMA200']:
        print(f"Sell {ticker} stock: Closing price is above 100 and 200 day SMA")
        
    #Display the opening, high, low, closing prices, 20 and 50 day Simple Moving Averages
    print(f"Stock: {ticker}")
    print(f"Latest Data Date: {curr_datetime}")
    print(f"Opening Price: {latest_data['Open']}")
    print(f"High Price: {latest_data['High']}")
    print(f"Low Price: {latest_data['Low']}")
    print(f"Closing Price: {latest_data['Close']}")
    print(f"100-Day SMA: {latest_data['SMA100']}")
    print(f"200-Day SMA: {latest_data['SMA200']}")
    
    # Extract the 'Close' column and reset the index
    stock_data = stock_data[['Close']].reset_index()
    # Rename the columns to match the Prophet input format
    stock_data = stock_data.rename(columns={'Date':'ds', 'Close': 'y'})
    model = Prophet(uncertainty_samples=10000,daily_seasonality=True)
    model.fit(stock_data)
    future = model.make_future_dataframe(periods=365)
    forecast = model.predict(future)
    
    # Plot the forecast using matplotlib
    plt.figure()
    plt.plot(forecast['ds'], forecast['yhat'], label=ticker)
    plt.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], alpha=0.3)
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.title(f"Forecast for {ticker}")
    plt.legend()
    plt.show()
```

#### The report concludes by summarizing the key findings of the time series analysis. This includes the trading signals derived from the relationship between the closing price and the moving averages, insights from the exploratory data analysis, and the forecasted stock prices obtained from the Prophet model.
